```python
import pandas as pd
import matplotlib.pyplot as plt
from numpy.random import randint
import folium as fo
from folium.plugins import HeatMap
import requests

pokemon = pd.read_csv('pokemon-spawns.csv')
pokemon.head()
```

```python
pokemon.info()
```

```python
for i in pokemon:
    print(i)
    print(pokemon[i].unique())
```

```python
values = randint(0, len(pokemon), 100)

myMap = fo.Map(
    location = [37.6, -122.2],
    tiles = 'OpenStreetMap',
    zoom_start = 9.5,
)

for i in values:
    image_pokemon = 'https://pokeres.bastionbot.org/images/pokemon/' + str(pokemon.iloc[i]['num']) + '.png'
    url = 'https://pokeapi.co/api/v2/pokemon/' + pokemon.iloc[1]['name'].lower()
    get_poke = requests.get(url)
    data_pokemon = get_poke.json()
    height = data_pokemon['height']
    weight = data_pokemon['weight']
    order = data_pokemon['order']
    
    fo.Marker(
        location=[pokemon.iloc[i]['lat'], pokemon.iloc[i]['lng']],
        popup = '<h4>' + pokemon.iloc[i]['name'] +'</h4>\
                <img src=' + image_pokemon + ' width="140%" height="auto">\
                <table>\
                  <tr>\
                    <th>Criteria</th>\
                    <th></th>\
                  </tr>\
                  <tr>\
                    <td><b>ID</b></td>\
                    <td><center>' + str(pokemon.iloc[i]['num']) + '</center></td>\
                  </tr>\
                  <tr>\
                    <td><b>Height</b></td>\
                    <td><center>' + str(height) +'</center></td>\
                  </tr>\
                  <tr>\
                    <td><b>Weight</b></td>\
                    <td><center>' + str(weight) +'</center></td>\
                  </tr>\
                  <tr>\
                    <td><b>Order</b></td>\
                    <td><center>' + str(order) +'</center></td>\
                  </tr>\
                </table>\
                ',
        tooltip = '<center>' + pokemon.iloc[i]['name'] + '</center>' + '<img src=' + image_pokemon + ' width=120% height=auto>',
        icon = fo.features.CustomIcon(icon_image=image_pokemon, icon_size=45)
    ).add_to(myMap)
    
loc = 'Pokemon Map in San Francisco Bay Area'
title_html = '''
             <h2 align="center" style="color:#72b591"><b>{}</b></h2>
             '''.format(loc)   
myMap.get_root().html.add_child(fo.Element(title_html))
myMap
```
